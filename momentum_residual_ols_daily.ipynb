```python
import statsmodels.api as sm
from statsmodels.sandbox.regression.predstd import wls_prediction_std
from sklearn.datasets import make_regression
import time
import multiprocessing

##200번째 까지는 저장완료

def residual_momentum_ols():
    col_list = return_daily.columns[200:]
    date_lis = three_factor_daily['2014':].index
    # date_lis = three_factor_daily

    start_time = time.time()
    print(len(col_list))
    ols_book = dict()
    
    for k,col in enumerate(col_list):
        
        if k % 10 == 0:
            print(str(k)+'번째'+str(col),(time.time()-start_time)/60*60)
            
        residualmomentum_ols_result_df = pd.DataFrame(columns=['Symbol','Date','Residual6','Residual12'])
        residual6_list = []
        residual12_list=[]
        date_list = []
        symbol_list = []


        for i,date in enumerate(date_lis):
            try:
                start_date = date_lis[i]
                end_date = date_lis[i+719]
                reval_date = str(pd.date_range(date_lis[i+719], periods=3)[-1]).split(' ')[0]

                temp = three_factor_daily[start_date:end_date]
                temp['Return']= return_daily[start_date:end_date][col].values
                temp['Ret-Rf'] = temp['Return']-temp['Rf']

                if len(temp.dropna())<=700:
                    pass

                else:
                    ols_temp = temp[['Ret-Rf','Mkt-Rf','SMB','HML']]
                    ols_temp = sm.add_constant(ols_temp)
                    x_train = ols_temp[['const','Mkt-Rf','SMB','HML']]
                    y_train = ols_temp['Ret-Rf']
                    result = sm.OLS(y_train,x_train).fit()

                    six = temp['Ret-Rf']-result.predict()
                    twelve = temp['Ret-Rf']-result.predict()

                    residual6_list.append(np.mean(six[-140:-20])/np.std(six[-140:-20]))
                    residual12_list.append(np.mean(twelve[-260:-20])/np.std(twelve[-260:-20]))
                    date_list.append(reval_date)
                    symbol_list.append(col)

            except:
                break

        residualmomentum_ols_result_df['Symbol']=symbol_list
        residualmomentum_ols_result_df['Date']= date_list
        residualmomentum_ols_result_df['Residual6']=residual6_list
        residualmomentum_ols_result_df['Residual12']=residual12_list
        ols_book[col] = residualmomentum_ols_result_df
        
    temp = pd.DataFrame()
    for key in ols_book.keys():
        temp =pd.concat([temp,ols_book[key]])
        
    temp.to_csv('200에서 부터 ols result.csv')

residual_momentum_ols()
```
